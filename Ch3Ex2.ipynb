```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([88,225,365,687,4329,10753,30660,60150])
y = np.array([57.9,108.2,149.6,227.9,778.1,1428.2,2837.9,4488.9])
```

```python
plt.scatter(x,y, color = 'b')
plt.grid()
plt.show()
```

```python
x_new = np.log(x)
y_new = np.log(y)
```

```python
coeff = np.polyfit(x_new,y_new,1)
poly = np.poly1d(coeff)
x_plot = np.linspace(min(x_new), max(x_new), 100)
y_plot = poly(x_plot)


plt.scatter(x_new, y_new, color = 'b')
plt.plot(x_plot, y_plot, color = 'r')
plt.grid()
plt.annotate(poly, xy=(0.05, 0.9), xycoords="axes fraction", fontsize=10)
```

```python
# Equation
def equation(x):
    return 2.94 * (x ** 0.666)

# Generate x values for plotting the equation
x_values = np.linspace(min(x), max(x), 100)

# Evaluate the equation for each x value
y_values = equation(x_values)

# Create the plot
plt.plot(x_values, y_values, color='b')
plt.scatter(x, y, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
